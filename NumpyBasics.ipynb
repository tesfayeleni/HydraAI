```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,3,2,4], '-b')
plt.show()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax1.plot([1, 3, 2, 4], '-b') # a simple line of points is given
ax1.set_title("Line plot")

ax2=fig.add_subplot(2,2,2)
ax2.scatter([1,2,3,4], [10,20,25,30], color='r')
ax2.set_title("Scatter plot")

ax3=fig.add_subplot(2,2,3)
x = np.linspace(0, 2*np.pi, 100) 
ax3.plot(x, np.sin(x))   # a sine fucntio is passed on to this! S
ax3.set_title("Sine curve")

ax4=fig.add_subplot(2,2,4)
ax4.bar([1,2,3,4], [5,7,3,8], color='g')
ax4.set_title("Bar chart")

plt.tight_layout()
plt.show()
```
